```python
#importamos las librerías a utilizar

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#cargamos el dataset

df_venezuelans = pd.read_csv('venezuelan_migrants(2000-2021).csv', index_col = 0)
print(df_venezuelans)
```

```python
#revisamos/verificamos el tipo de datos de cada columna

df_venezuelans.dtypes
```

```python
#observamos los primeros 10 registros, de los 521184 que contiene el dataset

df_venezuelans.head(10)
```

```python
#verificamos que solo estén incluidos migrantes de Venezuela

df_venezuelans.PAÍS.unique()
```

```python
#eliminamos entonces el país, ya que no aporta información al análisis

df_venezuelans = df_venezuelans.drop(columns = ['PAÍS'])

#verificamos que se haya eliminado dicha columna

df_venezuelans.head()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas otorgadas por año

df_venezuelans['AÑO'].value_counts()
```

```python
#observemos gráficamente la cantidad de visas temporarias
#otorgadas a ciudadanos venezolanos para residir en Chile, 
#desde el año 2000 al 2021

plt.figure(figsize = (8, 6))
sns.countplotx = 'AÑO', data = df_venezuelans, color = 'green')
plt.grid()
plt.title('Visas otorgadas por año', fontsize = 15)
plt.xlabel('Año', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas otorgadas según beneficio

df_venezuelans['BENEFICIO_AGRUPADO'].value_counts()
```

```python
#observemos gráficamente la cantidad de visas otorgadas
#a ciudadanos venezolanos para residir en Chile, 
#separadas por el motivo (beneficio) presentado, 
#respecto en todo el período estudiado (2000-2021)

plt.figure(figsize = (6, 4))
sns.countplot(x = 'BENEFICIO_AGRUPADO', data = df_venezuelans)
plt.grid()
plt.title('Visas otorgadas por beneficio', fontsize = 15)
plt.xlabel('Beneficio para visa', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas otorgadas según sexo, durante todo el período

df_venezuelans['SEXO'].value_counts()
```

```python
#graficamos la cantidad de visas otorgadas
#a ciudadanos venezolanos para residir en Chile, 
#separadas por sexo, y en cada año, desde el 2000 al 2021

plt.figure(figsize = (8, 6))
sns.countplot(x = 'AÑO', hue = 'SEXO', data = df_venezuelans)
plt.grid()
plt.legend(['Hombre','Mujer'])
plt.xlabel('Año', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtenemos datos numéricos respecto a la edad
#de las personas a las que se les ha otorgado una visa temporaria
#promedio, máximo, mínimo, y cuartiles

print(df_venezuelans['EDAD'].mean())
print(df_venezuelans['EDAD'].max())
print(df_venezuelans['EDAD'].min())
print(df_venezuelans['EDAD'].quantile([0, .25, .5, .75, 1]))
print(df_venezuelans['EDAD'].quantile([0, .14, .5, .97, 1]))
```

```python
#cercs del 83% de las visas han sido otorgadas a personas entre 18 y 60 años
#visualicemos el resultado anterior por medio de un gráfico
#que muestre la cantidad de venezolanos que le ha sido aprobada la
#visa de residencia temporaria en los años comprendidos del 2000 al 2021
#según edad

plt.figure(figsize = (20, 6))
sns.countplot(x = 'EDAD', data = df_venezuelans)
plt.grid()
plt.xlabel('Edad', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas otorgadas según titularidad

values = df_venezuelans['TIT_DEP'].value_counts()
values
```

```python
#observemos los porcentajes de visas otorgadas en el período estudiado
#según la titularidad (titular, dependiente o Sin información)

names = ['Titular', 'Dependiente', 'Sin información']
plt.pie(values,
        labels = names,
        shadow = True,
        explode = [0, 0.2, 0.4],
        autopct ='%1.1f%%')

plt.title("Porcentaje de visas otorgadas según titularidad")
plt.show()
```

```python
#a continuación, graficamos la cantidad de visas otorgadas
#a ciudadanos venezolanos para residir en Chile, 
#en relación a si el beneficiario es titular o dependiente (carga de un titular),
#y por cada año incluido en el estudio, del 2000 al 2021

plt.figure(figsize = (14, 6))
sns.countplot(x = 'AÑO', hue = 'TIT_DEP', data = df_venezuelans)
plt.grid()
plt.legend(['Titular','Dependiente','Sin informacion'])
plt.xlabel('Año', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas otorgadas según estudios completados

values = df_venezuelans['ESTUDIOS2'].value_counts()
values
```

```python
#observemos los porcentajes de las visas otorgadas en el período
#según estudios del venezolano migrante

names = ['Secundaria', 'Universitario', 'No Informa', 'Técnico', 'Prebásico']
plt.pie(values,
        labels = names,
        shadow = True,
        explode = [0.1, 0, 0.1, 0.2, 0.3],
        autopct ='%1.1f%%')

plt.title("Porcentaje de visas otorgadas según nivel de estudios")
plt.show()
```

```python
#separemos los registros de migrantes con estudios unirseritarios
#agrupados por año

df_university_studies = df_venezuelans[df_venezuelans['ESTUDIOS2'] == 'Universitario'].groupby('AÑO').count()
df_university_studies.head(22)
```

```python
#ahora realizamos la separación de los registros de migrantes con estudios técnicos
#agrupados por año

df_technical_studies = df_venezuelans[df_venezuelans['ESTUDIOS2'] == 'Técnico'].groupby('AÑO').count()
df_technical_studies.head(22)
```

```python
#lo mismo que lo anterior, pero para migrantes con estudios secundarios
#agrupados por año

df_high_school_studies = df_venezuelans[df_venezuelans['ESTUDIOS2'] == 'Secundaria'].groupby('AÑO').count()
df_high_school_studies.head(22)
```

```python
#ahora los migrantes con estudios prebásica, agrupados por año

df_preschool_studies = df_venezuelans[df_venezuelans['ESTUDIOS2'] == 'Prebásico'].groupby('AÑO').count()
df_preschool_studies.head(22)
```

```python
#grafiquemos las tendencias de visas otorgadas por cada uno de los
#niveles educacionales informados, respecto a cada año desde el 2000 al 2021

years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010',
       '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
df_studies = pd.DataFrame({'Año':years,
                           'Universitario':df_university_studies['ESTUDIOS2'],
                           'Técnico':df_technical_studies['ESTUDIOS2'],
                           'Secundaria':  df_high_school_studies['ESTUDIOS2'],
                           'Básica':df_preschool_studies['ESTUDIOS2']})
df_studies = df_studies.set_index('Año')
fig, ax = plt.subplots()
fig.set_size_inches(14, 6)
df_studies.plot(ax = ax)
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas temporarias otorgadas a venezolanos
#en función de la ocupación

occupation = df_venezuelans['OCUPACION'].value_counts()
occupation
```

```python
#ahora observemos los datos gráficamente, la cantidad de visas otorgadas
#a ciudadanos venezolanos para residir en Chile, 
#separado por la ocupación que reportaron realizar en el país,
#y para todo el período del estudio (del 2000 al 2021)

plt.figure(figsize = (10, 6))
sns.countplot(x = 'OCUPACION', data = df_venezuelans)
plt.grid()
plt.title('Visas otorgadas por ocupación reportada', fontsize =15)
plt.xlabel('Ocupación', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#revisemos los porcentajes de migrantes venezolanos
#que ejercían actividad remunerada en Chile al momento de solicitar la visa, 
#respecto a aquellos que no la ejercían

paid_work = occupation[0] + occupation[5] + occupation[6]
unpaid_work = occupation[1] + occupation[2] + occupation[3] + occupation[7]
other_activity = occupation[4]

values = [paid_work, unpaid_work, other_activity]
names = ['Trabajo remunerado', 'Trabajo no remunerado', 'Otra actividad']
plt.pie(values,
        labels = names,
        shadow = True,
        explode = [0.1, 0.15, 0.2],
        autopct ='%1.1f%%')

plt.title("Porcentaje de visas otorgadas según actividad")
plt.show()
```

```python
#obtenemos numéricamente, y con orden descendente,
#la cantidad de visas temporarias otorgadas a venezolanos
#en cada región de Chile, y para todo el período de estudio

df_venezuelans['REGIÓN2'].value_counts()
```

```python
#ahora observemos los datos gráficamente, la cantidad de visas otorgadas
#a ciudadanos venezolanos para residir en Chile, 
#separado por la regiones del país,
#y para todo el período del estudio (del 2000 al 2021)

plt.figure(figsize = (12, 6))
sns.countplot(x = 'REGIÓN2', data = df_venezuelans)
plt.grid()
plt.title('Visas otorgadas por Región de Chile', fontsize =15)
plt.xlabel('Región', fontsize = 15)
plt.xticks(fontsize = 10) 
plt.ylabel('Cantidad de visas otorgadas', fontsize = 15)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#ahora separemos los registros de los migrantes que residen en la RM
#para observar su distribución geográfica por comuna dentro de la capital

df_venezuelans_rm = df_venezuelans[df_venezuelans['REGIÓN'] == 'Metropolitana de Santiago']
df_venezuelans_rm.head()
```

```python
#verificamos que la cantidad de registros coincide con el número
#obtenido previamente respecto a los migrantes residenres en la capital

df_venezuelans_rm.shape
```

```python
#observemos los datos gráficamente, la cantidad de visas otorgadas
#a ciudadanos venezolanos según lugar de residecnia en comunas de la capital, 
#y para todo el período del estudio (del 2000 al 2021)

plt.figure(figsize = (16, 8))
sns.countplot(x = 'COMUNA', data = df_venezuelans_rm)
plt.grid()
plt.title('Visas otorgadas según comuna deresidencia en la RM', fontsize = 15)
plt.xlabel('Comuna', fontsize = 15)
plt.xticks(fontsize = 10, rotation = 90) 
plt.ylabel('Cantidad de visas otorgadas en el período 2000-2021', fontsize = 17)
plt.yticks(fontsize = 10) 
plt.show()
```

```python
#obtengamos el orden según la cantidad de migrantes por comuna
#de la comunas con mayor número de migrantes venezolanos que han residido
#con visa temporaria, a las de menos número de migrantes

df_venezuelans_rm['COMUNA'].value_counts()
```

CONCLUSIONES:

Del análisis se puede señalar que a la mayoría de los migrantes de nacionalidad venezolana se les otorgó la visa temporaria entre los años 2016 y 2021, siendo 2019 el año en que más se otorgaron visas, incluso por encima del año 2018 en el que se desarrolló en Chile un proceso de regularización de migrantes. El mismo año 2019, el Gobierno de Perú implementó una visa para todos los venezolanos que fueran a ingresar a ese país. Lo anterior, originó un aumento de migrantes venezolanos en el norte de Chile, impulsando al Gobierno chileno a ejecutar medidas similares al gobierno de Perú; específicamente, Chile implementó la obligatoriedad de una visa consular a todos los turistas venezolanos que quisieran ingresar a territorio nacional. 

Respecto al sexo de los migrantes venezolanos en Chile, en el período analizado no existen diferencias significativas en las visas otorgadas a los solicitantes, encontrándose cerca del 50% tanto hombres como mujeres. En torno a la titularidad, al 84.7% de los solicitantes que se les entregó visa fue con la condición de titular; aunque se debe tener en cuenta que hay un 4.9% del que no se tiene información, por lo que este porcentaje podría ser mayor. El nivel educacional informado, es en su gran mayoría Universitaria y Secundario, prevaleciendo esta última por sobre la primera en los últimos años del período analizado.

Adicionalmente, se puede establecer que la mayoría de los migrantes venezolanos a lo que se les otorgó la visa se encontraban ubicados en el centro del país, permitiendo esta información coadyuvar en la planificación de los recursos regionales y municipales. La Región Metropolitana lideró con un 66,30%, seguido por la Región de Valparaíso con sólo el 8,21%, por Biobío con 4,05%, y por Maule con 3,87%. En cuanto a las comunas, la de Santiago encabeza el listado con un 23,3% del total del país (con 121428 migrantes), seguida por Estación Central con 26850 migrantes, e Independencia con 24550.

En cuanto a la ocupación, es importante señalar que la mayoría de las personas venezolanas a las que se les otorgó la visa temporaria, se encontraban en Chile con la intención de ejercer alguna actividad remunerada, lo cual se evidencia en el hecho que el 27% de las visas fueron del tipo “Sujeta a contrato” y el 24,31% fueron del tipo “Motivo Laborales”. En decir, más del 50% del total de visas corresponde a personas que ya tenían un contrato de trabajo al momento de solicitarla. En este mismo sentido, destacar que la mayoría de los migrantes venezolanos en el período de estudio, residían en el país con un empleo, realizando el 76.2% de ellos un trabajo remunerado: como empleado, emprendedor, obrero o jornalero. A lo anterior se suma que el 25,64% de las visas fueron del tipo “Profesional”, la cual habilita a la persona a ejercer cualquier actividad lícita de acuerdo con el área de especialidad de su título profesional; y que cerca del 83% de los migrantes venezolanos se encontraban en edades entre los 18 y 60 años (edad laboralmente activa), siendo la media de edad 29 años. Con todo ello, se evidencia el aporte como fuerza laboral calificada, y por ende, aporte económico y social, que ha constituido la migración venezolana en Chile, especialmente en los últimos años, período en el que esta ha crecido.
