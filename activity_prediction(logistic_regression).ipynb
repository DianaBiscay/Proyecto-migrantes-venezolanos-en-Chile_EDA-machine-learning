PROYECTO: MIGRANTES VENEZOLANOS EN CHILE

DATOS: archivo venezuelan_migrants(2000-2021).csv (producto del análisis inicial de los datos).

OBJETIVO GENERAL: 

1. Predecir, a partir de los datos de registro de solicitud de visa temporaria, si el migrante venezolano ejerce actividad remunerada en el territorio chileno.

OBJETIVOS ESPECÍFICOS:

1.1 Utilizar un algoritmo de regresión logística (aprendizaje de máquina supervisado, a partir de los datos de migrantes venezolanos a los que fue otorgada una visa temporaria durante el período 2000-2021), que permita predecir si el migrante venezolano ejerce actividad remunerada en territorio chileno.

1.2 Utilizar el mismo algoritmo para establecer, con cierta probabilidad, si los migrantes que declararon en OCUPACIÓN "Otras Actividades", realizaban actividad remunerada, o no remunerada, en el país.

DEFINICIÓN DE MIGRANTE VENEZOLANO EN CHILE (para este proyecto): Aquel ciudadano venezolano al que una autoridad estatal chilena le aprobó una visa de residencia temporaria, que le permite permanecer de manera legal en territorio nacional por un período de tiempo definido.

Notas respecto al análisis inicial de los datos (procesos revios):

- Los datos fueron obtenidos de la página oficial del Servicio Nacional de Migraciones de Chile el 03 de febrero de 2022.
- Se trabajó con los datos relativos a las visas de residencia temporaria otorgadas durante el período 2000-2021.
- El análisis exploratorio permitió concluir que el 76.2% de los migrantes venezolanos que se los otorgó la visa temporaria en dicho período reportaron encontrarse realizando un un trabajo remunerado: como Empleado, Emprendedor, Obrero o Jornalero.

```python
#importamos las librerías a utilizar

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import joblib
import collections

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#cargamos el dataset

df_venezuelans = pd.read_csv('venezuelan_migrants(2000-2021).csv', index_col = 0)
print(df_venezuelans)
```

```python
#observamos los primeros 10 registros

df_venezuelans.head(10)
```

```python
df_venezuelans['COMUNA'].value_counts()
```

```python
#eliminemos las columnas que no aportan al análisis
#SEXO, variable categórica que fue transformada a numérica en la columna SEXO2
#TIT_DEP, variable categórica que fue transformada a numérica en la columna TIT_DEP2
#PAÍS, es el mismo para todos los registros
#NACIMIENTO y AÑO_NAC quedan representados en la columna EDAD, entendiendo que el día y mes de nacimiento no aporta al análisis
#en la localidad se incorporará únicamente la REGIÓN, siendo eliminadas PROVINCIA, COMUNA y REGIÓN2, en aras de disminuir la cantidad de variables en torno a este ítem
#AUTORIDAD que otorga la visa, ya que no está relacionado con la actividad que ejerce el migrante
#finalmente, AÑO y MES en que se otorgó la visa también se eliminará, ya que el objetivo principal de la predicción es ser utilizada en tiempo futuro.

df_venezuelans = df_venezuelans.drop(columns = ['SEXO', 'PAÍS', 'NACIMIENTO', 'PROVINCIA', 'TIT_DEP', 'COMUNA', 'AUTORIDAD', 'AÑO', 'MES', 'AÑO_NAC', 'REGIÓN2'])
df_venezuelans.head()
```

```python
#revisemos los diversos valores que puede tomar la variablle OCUPACIÓN

df_venezuelans['OCUPACION'].value_counts()
```

```python
#sustituyamos las categorías de OCUPACIÓN por valores numéricos que representen Otras Actividades (2), actividad remunerada (1), no remunerada (0)
#de ellas, constituyen actividades remuneradas: Empleado, Emprendedor, y Obrero o Jornalero
#constituyen actividades no remuneradas: Estudiante, Sin ocupación, Dueña de Casa, y Jubilado o Pensionado

df_venezuelans = df_venezuelans.replace({'Otras Actividades': 2, 'Dueña De Casa': 0, 'Empleado': 1, 'Emprendedor' : 1, 'Estudiante' : 0, 'Jubilado O Pensionado' : 0, 'Obrero O Jornalero' : 1, 'Sin ocupacion' : 0})
```

```python
#verifiquemos que se sustituyó por 3 categorías numércicas

df_venezuelans['OCUPACION'].value_counts()
```

```python
#veamos el porcentaje de registros de cada actividad

df_venezuelans.OCUPACION.value_counts(True)
```

```python
#el 76.2% de los registros constituyen migrantes que reportaron actividad remunerada en el período estudiado. 
#el porcentaje coincide con el inicial (EDA, proceso previo)
```

```python
#ahora trabajaremos con las columnas categóricas para codificarlas (llevarlas a valores numéricos)

categorical_variables = df_venezuelans.select_dtypes([object])
categorical_variables
```

```python
categorical_variables = pd.get_dummies(categorical_variables)
categorical_variables
```

```python
#revisemos las variables numéricas

numeric_variables = df_venezuelans.select_dtypes([int, float])
numeric_variables
```

```python
#concatenamos ahora todas las columnas, siendo numéricas

df_final = pd.concat([numeric_variables, categorical_variables], axis = 1)
df_final
```

```python
#el valor Otras Actividades, no puede definirse como actividades remuneradas o no, ya que no se cuenta con mayor información
#por ello, utilizaremos más adelante estos registros para predecir si la actividad resportada en esta categoría
#es remunerada o no, con cierta probabilidad.
#por ello, reservaremos estos registros y los eliminaremos del dataset para el proceso de entrenamiento del algoritmo

indexNames = df_final[df_final['OCUPACION'] == 2].index
df_other_activities = df_final.loc[indexNames, :]
df_final.drop(indexNames, inplace = True)
df_final.shape
```

```python
df_other_activities.head()
```

```python
df_other_activities.shape
```

```python
#al ser eliminados los registros con ocupación "Otras Actividades"
#reseteamos los índices del dataframe original

df_final.reset_index(inplace = True)
df_final = df_final.drop(["index"], axis = 1)
df_final
```

```python
#verificamos que en el dataframe original no queden registros con "Otras Actividades" (2)
#en la columna OCUPACIÓN

df_final.groupby('OCUPACION').count()
```

```python
#ahora vamos a dividir los datos en datos de entrenamiento y datos de prueba
#utilizaremos el 70% de los datos para entrenamiento y el 30% de los datos para prueba

X_train, X_test, y_train, y_test = train_test_split(df_final.drop("OCUPACION", axis = 1), 
                                                    df_final.OCUPACION, 
                                                    test_size = 0.3,
                                                    random_state = 7)
X_train.shape
```

```python
#utilizaremos un modelo de aprendizaje de máquina supervisado: regresión logística
#este algoritmo permite predecir valores binarios, en nuestro caso representados por Actividad Remunerada y Actividad no Remunerada

LogisticRegression?
```

```python
#creamos el modelo y lo etrenamos

regression_model = LogisticRegression(fit_intercept = True, penalty = 'l2', tol = 1e-6, C = 0.8, solver = 'lbfgs', max_iter = 1000,
                       warm_start = True)
regression_model.fit(X_train, y_train)

#guardamos las predicciones del modelo (y sus probabilidades asociadas) para los datos de prueba

predictions = regression_model.predict(X_test)
real_values = y_test
probabilistic_predictions = regression_model.predict_proba(X_test)
```
